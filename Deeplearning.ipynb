```python
import numpy as np
```

```python
def relu(x):
  return np.maximum(0,x)
```

```python
relu(10)
```

```python
relu(-10)
```

```python
relu(np.array([-1,0,1,2,-5,11]))
```

```python
def leky_relu(x, alpha=0.01):
  return np.where(x>0, x, x*alpha)
```

```python
leky_relu(-10)
```

```python
leky_relu(np.array([-3,-2,0,2,4]))
```

```python
def elu(x, alpha=1.0):
  return np.where(x>0,x,alpha*(np.exp(x)-1))
```

```python
elu(5)
```

```python
elu(np.array([-3,-2,0,2,4]))
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
sigmoid(np.array([-3,-2,0,2,4]))
```

```python
def softmax(x):
  e_x=np.exp(x-np.max(x))
  return e_x/e_x.sum(axis=0)
```

```python
y=softmax(np.array([-1,0,3]))
```

```python
def sigmod(z):
  return 1/(1+np.exp(-z))
```

```python
def feedforward(weight,bias,input):
  total=sum(w * x for w,x in zip(weight,input))+bias
  return sigmod(total)
```

```python
def classify(output,threshold=0.5):
  return 1 if output>threshold else 0
```

```python
weight=[0,1]
bias=4
input=[2,3]
output=feedforward(weight,bias,input)
print(f'output of the neorun for input {input} is {output}')
```

```python
class_level=classify(output)
print(f'class level is {class_level}')
```
