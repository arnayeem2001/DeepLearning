```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
input_dim = 3
# hidden dimension (h_prev এর size)
hidden_dim = 4

# weight matrices
wf = np.random.randn(hidden_dim + input_dim, hidden_dim)
wi = np.random.randn(hidden_dim + input_dim, hidden_dim)
wc = np.random.randn(hidden_dim + input_dim, hidden_dim)
wo = np.random.randn(hidden_dim + input_dim, hidden_dim)

# bias vectors
bf = np.zeros((1, hidden_dim))
bi = np.zeros((1, hidden_dim))
bc = np.zeros((1, hidden_dim))
bo = np.zeros((1, hidden_dim))
```

```python
h_prev = np.zeros((1, hidden_dim))
x_t = np.random.randn(1, input_dim)

concat = np.concatenate((h_prev, x_t), axis=1)
```

```python
# LSTM equations
ft = sigmoid(np.dot(concat, wf) + bf)
it = sigmoid(np.dot(concat, wi) + bi)
cct = np.tanh(np.dot(concat, wc) + bc)
c_next = ft * h_prev + it * cct   # এখানে c_prev ব্যবহার করা উচিত
ot = sigmoid(np.dot(concat, wo) + bo)
h_next = ot * np.tanh(c_next)
```
