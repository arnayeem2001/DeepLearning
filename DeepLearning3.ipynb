```python
one_geo = OneHotEncoder(drop='first')

# ধরো তোমার DataFrame এ কলামের নাম 'Treatment_Type'
geo_encoded = one_geo.fit_transform(data[['Platelet Count( (/cumm)']])

# DataFrame এ রূপান্তর করো
geo_encoded_df = pd.DataFrame(
    geo_encoded.toarray(),
    columns=one_geo.get_feature_names_out(['Platelet Count( (/cumm)'])
)

print(geo_encoded_df.head())
```

```python
data=pd.concat([data.drop(columns=['Platelet Count( (/cumm)']),geo_encoded_df],axis=1)
```

```python
le_Treatment_Outcome=LabelEncoder()
data['Treatment_Outcome']=le_Treatment_Outcome.fit_transform(data['Treatment_Outcome'])
```

```python
le_Treatment_Outcome=LabelEncoder()
data['Diagnosis_Result']=le_Treatment_Outcome.fit_transform(data['Diagnosis_Result'])
```

```python
le_Treatment_Outcome=LabelEncoder()
data['Comments']=le_Treatment_Outcome.fit_transform(data['Comments'])
```

```python
data.info()
```

```python
data.info()
```

```python
X=data.drop(columns=['Treatment_Outcome'])
y=data['Treatment_Outcome']
```

```python
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
data.describe()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = Sequential([
    Dense(units=32, activation='relu', input_dim=X_train.shape[1]),
    Dense(units=16, activation='relu'),
    Dense(units=8, activation='relu'),
    Dense(units=1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=100, validation_data=(X_test, y_test))
```

```python
y_pred=(model.predict(X_test)>0.5).astype(int)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import pandas as pd
import plotly.express as px
from sklearn.metrics import confusion_matrix

# ধরো y_test আর y_pred তোমার actual আর predicted labels
cm = confusion_matrix(y_test, y_pred)

# Class labels বের করো (unique values থেকে)
labels = [str(l) for l in sorted(set(y_test))]

# Confusion Matrix কে DataFrame বানাও
cm_df = pd.DataFrame(cm,
                     index=[f'Actual {l}' for l in labels],
                     columns=[f'Predicted {l}' for l in labels])

# Plotly দিয়ে visualize
fig = px.imshow(cm_df,
                text_auto=True,
                color_continuous_scale='Blues',
                title='Confusion Matrix')

fig.update_layout(
    title={'text': 'Confusion Matrix',
           'x': 0.5,
           'xanchor': 'center'},
    xaxis_title='Predicted Values',
    yaxis_title='Actual Values',
    coloraxis_showscale=True
)

fig.show()
```
