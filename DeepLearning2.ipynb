```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train.reshape(-1,28*28)/255.0
x_test=x_test.reshape(-1,28*28)/255.0
```

```python
model=Sequential([
    Dense(128,activation='relu',input_shape=(28*28,)),
    Dense(64,activation='relu'),
    Dense(32,activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(
    x_train,y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.25
)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred_class=np.argmax(y_pred,axis=1)
y_pred_class
```

```python
cm=confusion_matrix(y_test,y_pred_class)
print("Confusion Matrix:")
print(cm)
```

```python
display=ConfusionMatrixDisplay(confusion_matrix=cm)
display.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
model.save('mnist_model.h5')
```

```python
loaded_model=tf.keras.models.load_model('mnist_model.h5')
loaded_model
```

```python
random_index=np.random.randint(0,len(x_test))
random_image=x_test[random_index]
random_label=y_test[random_index]

random_image_reshaped=np.reshape(random_image,(1,28*28))
prediction=loaded_model.predict(random_image_reshaped)
prediction_label=np.argmax(prediction)

print(f"Random Index: {random_index}")
print(f"predicted Label: {prediction_label}")

plt.imshow(random_image.reshape(28,28),cmap='gray')
plt.title(f'True Label: {random_label},predicted Label: {prediction_label}')
plt.axis('off')
plt.show()
```
